```python
import pickle
import os
import numpy as np
from lmfit import  Model, Parameters
from ANEMO import read_edf
from ANEMO import ANEMO
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sujet = ['s01','s02','s03','s04','s05','s06','s07','s08','s09','s10','s11','s12']
```

```python
import glob
liste = []
for fname in glob.glob('../data/*pkl'):
    a = fname.split('/')[2].split('.')[0].split('_')
    liste.append(a)

BET = {}
for x in range(len(liste)) :
    if liste[x][0]=='bet' and liste[x][1] in sujet:
        a = '../data/%s_%s_%s_%s.pkl'%(liste[x][0], liste[x][1],liste[x][2],liste[x][3])
        with open(a, 'rb') as fichier :
            b = pickle.load(fichier, encoding='latin1')
            b['observer']=liste[x][1]
            BET[liste[x][1]] = b
```

```python
fig1, ax1 = plt.subplots(1,1, figsize=[15, 15/2])
N_cols = 4
fig, ax = plt.subplots(int(len(sujet)/N_cols), N_cols, figsize=[15, 15/2])

scaling_sujet = {}
scaling_full = []

a, b = 0, 0
bet_full = []
for i, x in enumerate(sujet) :
    ax[b][a].set_title(x)
    
    bet = []
    for block in range(BET[x]['N_blocks']) :
        bet.extend((BET[x]['results'][:,block]+1)/2)
    bet_full.extend(bet)
    scaling_sujet[x] = np.sort(bet)
    
    ax[b][a].plot(scaling_sujet[x], np.linspace(0, 1, len(scaling_sujet[x])), c='k')
    # a = i // N_cols
    # b = i % N_cols
    
    a = a+1
    if a > 3 :
        a = 0
        b = b+1

scaling_full = np.sort(bet_full)

ax1.set_title('Full')
ax1.plot(scaling_full, np.linspace(0,1,len(scaling_full)), c='k') 

fig.tight_layout()
fig1.tight_layout()

plt.show()
```

```python
c = ['k', 'grey', 'r']

new_proba_full = np.linspace(0,1,len(scaling_full))

for x in sujet :
    
    fig, ax = plt.subplots(3, 1, figsize=[15,15*1/2]) 
    
    new_proba_sujet = np.linspace(0, 1, len(scaling_sujet[x]))
    
    print(x)
    
    a = 0
    for block in range(3) :
        new_bet_sujet = []
        new_bet_full = []
        
        bet = (BET[x]['results'][:,block]+1)/2
        
        proba = BET[x]['p'][:,block, 1]
        
        for trial in range(len(bet)) :
            p = []
            for t in range(len(scaling_sujet[x])) :
                if scaling_sujet[x][t] == bet[trial] :
                    p.append(new_proba_sujet[t])
            new_bet_sujet.append(np.mean(p))
            
            p = []
            for t in range(len(scaling_full)) :
                if scaling_full[t] == bet[trial] :
                    p.append(new_proba_full[t])
            new_bet_full.append(np.mean(p))
        
        print('\nblock', block)
        print('bet : min = %0.3f -- max = %0.3f'%(min(bet), max(bet)))
        print('new_bet_sujet : min = %0.3f -- max = %0.3f'%(min(new_bet_sujet), max(new_bet_sujet)),
              'new_bet_full : min = %0.3f -- max = %0.3f'%(min(new_bet_full),max(new_bet_full)),
              sep='    ')
        
        
        ax[a].fill_between(range(len(proba)), np.zeros_like(proba), proba,
                                  lw=.5, alpha=0.2, facecolor='k', step='pre')

        
        ax[a].set_title(block)
        
        ax[a].plot(bet, c=c[block])
        ax[a].plot(new_bet_sujet, c=c[block], ls=':')
        ax[a].plot(new_bet_full, c=c[block], ls='--')
        
        a=a+1
    fig.tight_layout()
    plt.show()
```
