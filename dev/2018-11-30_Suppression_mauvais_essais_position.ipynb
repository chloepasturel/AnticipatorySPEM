# Figure

```python
def fct_fig(ax, data, exp, param_fit, trial, N_trials, block, opt_base) :
    
    A = ANEMO(exp)
    Fit = ANEMO.Fit(exp)
    
    trial_data = trial + N_trials*block

    arg = A.arg(data[trial_data], trial=trial, block=block)
    opt = opt_base.copy()
    opt.update(arg)

    data_x = A.data_deg(data=arg.data_x, **opt) #(arg.data_x - (arg.data_x[arg.StimulusOf-arg.t_0]))/arg.px_per_deg
    
    inde_v = Fit.generation_param_fit(equation=equation, **opt)[1]
    rv = {}
    onset  = arg.TargetOn - arg.t_0 
    for k in param_fit.keys() :
        if k in ['start_anti', 'latency'] :
            rv[k] = param_fit[k][block][trial] + onset
        else :
            rv[k] = param_fit[k][block][trial]
    
    rv['do_whitening'] = False    
    rv['nb_sacc'] = len(arg.saccades)
    rv['dir_target'] = arg.dir_target
    rv['t_0'] = arg.t_0
    rv['px_per_deg'] = arg.px_per_deg
    rv['before_sacc'] = 5
    rv['after_sacc'] = 15
    
    rv.update(inde_v)

    ax.set_ylabel('%s\nDistance (°)'%(trial+1), fontsize=t_label)

    start = arg.TargetOn
    trackertime_s = arg.trackertime - start
    TargetOn_s = arg.TargetOn - start
    TargetOff_s = arg.TargetOff - start

    ax.axis([TargetOn_s-700, TargetOff_s+10, -39.5/2, 39.5/2])
    ax = A.Plot.deco(exp, ax, **opt)

    fit = ANEMO.Equation.fct_position(**rv)
    ax.plot(trackertime_s[:-time_sup], fit[:-time_sup], color='r', linewidth=2)
    ax.plot(trackertime_s, data_x, color='k', alpha=0.4)

    y = 0
    for name in param_fit.keys() :
        value = param_fit[name][block][trial]
        if name in ['start_anti', 'latency'] :
            ax.bar(value, 80, bottom=-40, color='k', width=3, linewidth=0)
        ax.text((TargetOff_s-10), (35/2)+y, "%s: %0.3f"%(name, value) , color='k', fontsize=t_text, va='center', ha='right')
        y = y - 5/2

    ax.set_xlabel('Time (ms)', fontsize=t_label)
    return ax
```

```python
equation='fct_position'
time_sup = 280
t_label = 20
t_text = 14
fig_width = 12

opt_base = {'stop_search_misac': None, 'time_sup': 280,
            'param_fit': None, 'inde_vars': None,
            'step_fit':2, 'do_whitening': False,
            'before_sacc': 5, 'after_sacc': 15,
            't_label': t_label}    
```
