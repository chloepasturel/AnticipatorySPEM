```python
import pickle
import os
from ANEMO import read_edf
from ANEMO import ANEMO
import numpy as np
from lmfit import  Model, Parameters
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mode = 'eyeMvt'
from aSPEM import Analysis
e = Analysis(observer='s01', mode=mode, name_file_fit='fct_velocity_sigmo')
```

```python
Full = e.Full_list(modes_bcp=None)
```

```python
sujet = ['s01','s02','s03','s04','s05','s06','s07','s08','s09','s10','s11','s12']
```

```python
fig1, ax1 = plt.subplots(1,1, figsize=[15, 15/2])
fig, ax = plt.subplots(int(len(sujet)/4), 4, figsize=[15, 15/2])

scaling_sujet = {}
scaling_full = []

a, b = 0, 0
va_full = []
for x in sujet :
    ax[b][a].set_title(x)

    va_full.extend(list(Full['va'][Full.sujet==x]))
    scaling_sujet[x] = np.sort(list(Full['va'][Full.sujet==x]))
    
    ax[b][a].plot(scaling_sujet[x], np.linspace(0,1,len(scaling_sujet[x])), c='k')
    a = a+1
    if a > 3 :
        a = 0
        b = b+1

scaling_full = np.sort(va_full)

ax1.set_title('Full')
ax1.plot(scaling_full, np.linspace(0,1,len(scaling_full)), c='k') 

fig.tight_layout()
fig1.tight_layout()

plt.show()
```

```python
c = ['k', 'grey', 'r']

new_proba_full = np.linspace(0,1,len(scaling_full))

for x in sujet :
    
    fig, ax = plt.subplots(3, 1, figsize=[15,15*1/2])        
    new_proba_sujet = np.linspace(0, 1, len(scaling_sujet[x]))
    
    print(x)
    
    nb_trial = 0
    a = 0
    for block in range(3) :
        new_va_sujet = []
        new_va_full = []
        
        va = list(Full['va'][Full.sujet==x])
        
        va = va[nb_trial:200+nb_trial*block]
        for trial in range(len(va)) :
            p = []
            for t in range(len(scaling_sujet[x])) :
                if scaling_sujet[x][t] == va[trial] :
                    p.append(new_proba_sujet[t])
            new_va_sujet.append(np.mean(p))
            
            p = []
            for t in range(len(scaling_full)) :
                if scaling_full[t] == va[trial] :
                    p.append(new_proba_full[t])
            new_va_full.append(np.mean(p))
        
        print('\nblock', block)
        print('va : min = %0.3f -- max = %0.3f'%(min(va), max(va)))
        print('new_va_sujet : min = %0.3f -- max = %0.3f'%(min(new_va_sujet), max(new_va_sujet)),
              'new_va_full : min = %0.3f -- max = %0.3f'%(min(new_va_full),max(new_va_full)),
              sep='    ')
        
        ax1 = ax[a].twinx()
        ax1.plot(va, c=c[block])
        
        
        proba = list(Full['proba'][Full.sujet==x])
        proba = proba[nb_trial:200+nb_trial*block]
        ax[a].fill_between(range(len(va)), np.zeros_like(proba), proba,
                                  lw=.5, alpha=0.2, facecolor='k', step='pre')

        
        ax[a].plot(new_va_sujet, c=c[block], ls=':')
        ax[a].plot(new_va_full, c=c[block], ls='--')
        
        ax[a].set_title(block)
        
        a = a+1
        nb_trial = nb_trial+200
    fig.tight_layout()
    plt.show()
```
